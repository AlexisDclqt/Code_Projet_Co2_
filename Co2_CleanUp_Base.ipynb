```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_parquet(r"C:\Users\alexd\Desktop\WorkSpace\NoteBook_Jupyter\Projet_DA_DST_Co2\co2_cars.parquet")
```

```python
rows = data.shape[0]
columns = data.shape[1]
print(f'Le dataset contient {rows} lignes et {columns} colonnes.')

# Output : Le dataset contient 10779681 lignes et 40 colonnes.
```

```python
def missings(df):
    cols_with_na = 0

    for col in df.columns:
        n_missing = df[col].isna().sum()
        if n_missing > 0:
            cols_with_na += 1
            print(f'"{col}" : {n_missing} valeurs manquantes, soit {np.round((n_missing/df.shape[0])*100,2)}% des valeurs de la colonne')

    if cols_with_na == 0:
        print("Pas de valeurs manquantes")
    else:
        print(
            f"\nColonnes avec valeurs manquantes : {cols_with_na} / {df.shape[1]} "
            f"({np.round((cols_with_na / df.shape[1]) * 100, 2)}%)"
        )

missings(data)
```
